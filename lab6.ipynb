## warmup binary search


````python
def binary_search(sorted_list, target):

    # identify mid point

    total = len(sorted_list)

    midpoint = total // 2

    if total % 2 == 0:
        midpoint = total // 2

    if sorted_list[midpoint] == target:

        return midpoint

    if sorted_list[midpoint] > target:
        return binary_search(sorted_list[0:midpoint], target)

    if sorted_list[midpoint] < target:
        return midpoint + 1 + binary_search(sorted_list[midpoint + 1:], target)

    return -1

#test 

print(binary_search([1, 2, 3, 4, 5], 3)) # should return 2
````

````python
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class Tree:
    def __init__(self, root):
        self.root = root

    def bfs(self):
        """Return values in BFS (level-order)."""
        if self.root is None:
            return []

        order = []
        q = deque([self.root])
        while q:
            node = q.popleft()
            order.append(node.value)

            if node.left is not None:
                q.append(node.left)
            if node.right is not None:
                q.append(node.right)

        return order

    def dfs(self, order, root_start):

        if self.root is None:
            return []

        if root_start is None:
            return order
        
        q = deque([root_start])

        node = q.popleft()

        order.append(node.value)

        self.dfs(order, node.left)

        self.dfs(order, node.right)

        return(order)

````

## BFS can be used for shortest path problems, like you saw Thursday in lecture

Examine the following problem: Given an unweighted graph and two nodes `s` and `t`, find the minimum number of edges from `s` to `t` and the path taken.

BFS can be a good choice to solve this problem because it explores all neighbors of a node at the present depth, depth being the distance from the source node. 

````python
from collections import deque

def shortest_path(graph, start, target):
    if start == target:
        return 0

    # queue's should store the next nodes to process

    q = deque([start])
    visited = {start} # This keeps track of visited nodes. 
    distance = 0



    while q:
        # everything currently in q is exactly 'distance' steps from start
        level_count = len(q)

        for _ in range(level_count):

            node = q.popleft()
            # node is 'distance' steps from the start. It is not the target
            # check neighbors of node
            for neighbor in graph.get(node, []):
                if neighbor in visited:
                    continue
                if neighbor == target:
                    return distance + 1
                visited.add(neighbor)
                # Add the neighbor to the queue.
                # This added neighbor you know is now 'distance + 1' steps away from start
                q.append(neighbor)

        # finished this wave; next wave is one step farther
        distance += 1

    return -1
````

````python
graph = {
    "A": ["F", "B"],
    "B": ["A", "C", "D"],
    "C": ["B", "D", "E"],
    "D": ["E", "C", "B"],
    "E": ["D", "C"],
    "F": ["A"]
}

# Graph looks like (on board)

print(shortest_path(graph, "B", "E"))  # 2
````

### Number of Connected Components in a Undirected Graph

[Link](https://neetcode.io/problems/count-connected-components/question)

You can use BFS for these types of problems. You count how many times you have to start a traversal from an unvisted node. 

You start with the edges, which is a list of lists. You will definitely need to store the visited nodes. You can imagine everytime you begin a traversal on an unvisted node, thats one component

````python


class Solution:

    def BFS(self, start, graph, visited):

        q = deque([start])

        visited.add(start)

        while q:
            level_count = len(q)
            for _ in range(level_count):

                node = q.popleft()

                for neighbor in graph.get(node, []):
                    if neighbor in visited:
                        continue
                        
                    visited.add(neighbor)

                    q.append(neighbor)

        


    def countComponents(self, n: int, edges: list[list[int]]) -> int:

        # See board for schemantics

        # First, convert the edges to a dictionary

        graph = {i: [] for i in range(n)}

        for a, b in edges:
            graph[a].append(b)
            graph[b].append(a)
        
        visited = set()
        components = 0
        for node in range(n):
            if node in visited:
                continue

            # new node found
            # Do BFS from this unvisited node

            self.BFS(node, graph, visited)
            components += 1


        return(components)

    


        
````

````python
n = 5
edges = [[0,1],[1,2],[3,4]]
solution_comp = Solution()

solution_comp.countComponents(n, edges)
````

## DFS Leetcode Problem
# 207. Course Schedule
[Link](https://leetcode.com/problems/course-schedule/description/?envType=problem-list-v2&envId=depth-first-search)

There are a total of numCourses courses you have to take, labeled from `0` to `numCourses - 1`. You are given an array prerequisites where `prerequisites[i] = [ai, bi]` indicates that you must take course `bi` first if you want to take course `ai`.

For example, the pair `[0, 1]`, indicates that to take course `0` you have to first take course `1`.
Return true if you can finish all courses. Otherwise, return false.

Turn it into a graph problem. You're given `numCourses` number of edges, but they are directed now. 

You return true if it's a DAG! No cycles. 

How to detect cycles? Start with one node (class). Go down the depth of the node and see if you encounter a node you've already seen since you began on that node. 

Solution courtesy of [this article](https://cloudiosx.medium.com/207-course-schedule-dd5df9298716)

````python
class Solution(object):
    def DFS(self, graph, node, visiting):
        if node in visiting:
            return False  # cycle

        visiting.add(node)
        for prereq in graph[node]:
            if not self.DFS(graph, prereq, visiting):
                return False
        visiting.remove(node)
        return True

    def canFinish(self, numCourses, prerequisites):
        graph = {i: [] for i in range(numCourses)}
        for course, prereq in prerequisites:
            graph[course].append(prereq)

        for i in range(numCourses):
            if not self.DFS(graph, i, set()):  
                return False
        return True
````

````python
numCourses = 2
prerequisites = [[0,5],[1,2],[1,3],[2,4],[2,5]]

solution = Solution()
solution.canFinish(6,prerequisites)
````
